```python
blockheader = pd.read_csv("blockheader_data.csv")
blockheader = blockheader.iloc[1: , 2:17]
blockheader
```

```python
blockheader.columns = ['hash', 'confirmations', 'height', 'version',
                     'versionHex', 'merkleroot', 'time', 'mediantime',
                      'nonce', 'bits', 'difficulty', 'chainwork', 'nTx',
                      'previousblockhash', 'nextblockhash']
```

```python
blockheader
```

```python
blockheader["confirmations"] = blockheader["confirmations"].str.extract('(\d+)', expand=False)
blockheader["height"] = blockheader["height"].str.extract('(\d+)', expand=False)
blockheader["version"] = blockheader["version"].str.extract('(\d+)', expand=False)
blockheader["versionHex"] = blockheader["versionHex"].str.extract('(\d+)', expand=False)
blockheader["time"] = blockheader["time"].str.extract('(\d+)', expand=False)
blockheader["mediantime"] = blockheader["mediantime"].str.extract('(\d+)', expand=False)
blockheader["nonce"] = blockheader["nonce"].str.extract('(\d+)', expand=False)
blockheader["difficulty"] = blockheader["difficulty"].str.extract('(\d+)', expand=False)
blockheader["nTx"] = blockheader["nTx"].str.extract('(\d+)', expand=False)
```

```python
blockheader
```

```python
blockheader["hash"] = blockheader["hash"].str[11:75]
blockheader["merkleroot"] = blockheader["merkleroot"].str[17:81]
blockheader["bits"] = blockheader["bits"].str[11:19]
blockheader["chainwork"] = blockheader["chainwork"].str[16:80]
blockheader["previousblockhash"] = blockheader["previousblockhash"].str[24:88]
blockheader["nextblockhash"] = blockheader["nextblockhash"].str[20:84]
```

```python
blockheader.to_csv("clean_blockheader.csv")
```
